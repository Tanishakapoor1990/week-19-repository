For predicting response variables which are continous in nature  we have used the following: Linear Regression,
polynomial linear regression,multiple linear regression They are basically used for solving Regression problems.

Linear Regression :It's a statistical method that helps in finding relationship between an independent(X) 
and dependent variable(y),both of which are continous in nature.For eg. predicting salary based 
on employee ratings,predicting temperature of a given day.

It basically helps in finding linear relationship between X (input) and y(output). 
Therefore it is named Linear Regression.   

It follows this equation :
    y = mx + b 
    
Algorithm working :    
So first we define our X and y variables based on the  problem statement.
Then splitting it into train and test using train test split method
We further fit our model on X_train and y_train
After fitting the model it fits the best line to predict the value of y for a given value of X.The model gets the best
regression fit line by finding the best values for m and b .
m = slope 
b = intercept

Once we find the best values for slope and intercept, we get the best fit line. 
So when we are finally using our model for prediction, it will predict the value of y for the input value of X.

Multiple Linear regeression also works on similar principle as simple Linear regression .It is just an extension of Simple linear regression .
Here,two or more independent/predictors are used to predict variance in one dependent/response variable.The algorithm works the same way as it 
did for simple linear regression.

KNN(K-nearest neighbour): KNN algorithm assumes that similar things are near to each other.
The KNN Algorithm
1. Load the data
2. Initialize K to your chosen number of neighbors
3. For each point in the data
3.
    1 Calculate the distance(euclidean) between the query example and the actual example from the data.
    2 Add the distance and the index of the example to an ordered collection
4. Sort the ordered collection of distances and indices from smallest to largest (in ascending order) by the distances
5. Pick the first K entries that have the smallest distance calcluated using Euclidean distance.
6. Get the labels/response of the selected K entries
7. If regression, return the mean of the K labels
8. If classification, return the mode of the K labels

```python
For unsupervised models we have used Kmeans and hierarchial clustering.
Clustering :
In unsupervised learning we try to divide entire data into groups.These groups are called clusters and the process
of makimg clusters is known as Clustering.It tries to find similarity between points.

Kmeans Algorithm:
Kmeans algorithm is an iterative algorithm that tries to divide the dataset into Kpre-defined distinct 
non-overlapping (clusters) where each data point belongs to only one group.

Working of algorithm :
    1. Decide the number of clusters(k) to assign
    2. Randomly initialize k centroids
    3. Assign all the points to the nearest centroids
    4. Calculate the new cntroid (mean) of each cluster unti the centroid position stops changing

Evaluation Methods:
    Elbow method
    Silhouette analysis
    
For determinig the number of clusters we will use  Elbow method .

It gives an idea on what a good k number of clusters would be based on the sum of squared distance 
(SSE) between data points and their assigned clusters centroids. We choose  k at the spot where SSE starts to flatten out 
and forms an elbow.
When you increase k, SSE continues to decrease as more centroids are added.The distance from each points to it's 
closest centroid will decrease.


Hierrachial Clustering :
Hierarchical clustering algorithm is of two types:

    1) Agglomerative Hierarchical clustering algorithm(Bottom up) :
We start from individual clusters and and then gradually group clusters based on similarity in order to finally 
form one cluster.We are slowly building hierarchy of clusters.

    2) Divisive Hierarchical clustering algorithm(Top down ):We start with one cluster that has all the 
datapoints and then we keep dividing them into number of clusters.Hence, called top down approch as well.

Working of algorithm:(single-nearest distance or single linkage)
    
    1) We first create a Proximity matrix.This matrix tellsabout the distance between all data points.It's a square matrix as we are 
    calcualting the distance of each point from each of the other points. Diagonal elements are zero. 
    2) Assign all the dsta points to an individual cluster.
    3) Then , we look at the smallest distance in the proximity matrix and merge the points with the samallest distance.
    More the distance of the vertical lines in the dendogram more is the distance between those clusters.
    5)Set a threshold distance and draw a horizontal line.
    6) Number of vertical lines on the dendogram being intersected by the threshold line will give the number of clusters.
    
The process for finding optimal number of clusters can be determined using Elbow method.
    
```

## 4.Perform k-means on this dataset for customer segments. Customer segments help determine what types of people buy your product, which allows you to target more people like your usual customers. Should you look at all the data, or which subset of data should you use? What is the ideal number of clusters? Which approach did you use to find the ideal number of clusters and why?

I'm selecting a subset of my data for my analysis. I'm basically Clustering Customers Based on their purchasing behaviour. Type of customers based on the number of items they purchase and quantity.

For determining the optimal number of clusters i'm using elbow method. Elbow method give estimation of best value of k based on sum of squared errors between data points and their assigned clusters centroids. We choose k at the spot where SSE starts to flatten out and forms an elbow.

```python
# Applying K-Means Clustering Algorithm
# We use only 'Quantity','UnitPrice' columns to cluser 
X = df_cluster.iloc[:, [0, 1]].values
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X= sc_X.fit_transform(X)
```

```python
#Using the Elbow method to find the optical number of clusters
from sklearn.cluster import KMeans
wcss = [] #With in cluster sum of squers(Inertia)
'''
#n_clusters is no.of clusters given by this method,
#k-means++ is an random initialization methods for centriods to avoid random intialization trap,
#max_iter is max no of iterations defined when k-means is running
#n_init is no of times k-means will run with different initial centroids
'''
for i in range(1,11): #From 2-10 doing multiple random initializations can make a huge difference to find a better local optima
    kmeans = KMeans(n_clusters = i, init ='k-means++',max_iter=300,n_init=10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11) , wcss)
plt.title('The Elbow Method')
plt.xlabel('Number Of Customer Clusters(customer type clusters)')
plt.ylabel('With in cluster sum of squers(WCSS)')
plt.show()
```

```python
#From the plot we can see that at 3 distortion goes rapidly so n_clusters=3
# Fitting K-Means to the dataset
kmeans = KMeans(n_clusters = 3, init = 'k-means++')
y_kmeans = kmeans.fit_predict(X)

```

```python
# Visualising the clusters
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 50, c = 'red', label = 'Customer Type 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 50, c = 'blue', label = 'Customer Type 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 50, c = 'green', label = 'Customer Type 3')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 100, c = 'yellow', label = 'Centroids')
plt.title('Type Of Customers(customer type clusters)')
plt.xlabel('Number of items Purchased(Quantity)')
plt.ylabel('Product price per unit in sterling(Unit Price)')
plt.legend()
plt.show()
```
